```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as sts
import numpy as np
```

```python
# Study data files
mouse_metadata_path = "Pymaceuticals_data_Mouse_metadata.csv"
study_results_path = "Pymaceuticals_data_Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)
pd.set_option('display.max_rows', 2000)
```

```python
print("Here is the data about each mouse :") 
mouse_metadata
```

```python
print("Here are the study results on the mouse :") 
study_results
```

```python
# Combine the data into a single dataset
# Display the data table for preview
mousedata_and_results = pd.merge(mouse_metadata, study_results, how ='outer', on = 'Mouse ID')
pd.set_option('display.max_rows', 2000)
print("Here is the combined dataset :") 
mousedata_and_results
```

```python
#Before beginning the analysis, check the data for any mouse ID with duplicate time points and remove
#any data associated with that mouse ID
#Observations and Insights
```

```python
# Checking the number of mice.
number_of_mice = mousedata_and_results['Mouse ID'].nunique()

print(f" The number of unique Mice are : {number_of_mice}")
```

```python
# Getting the duplicate mice by ID number that shows up for Mouse ID and Timepoint.(returning dataframe) 
duplicateIDs = mousedata_and_results.loc[mousedata_and_results.duplicated(subset = ['Mouse ID', 'Timepoint']),['Mouse ID','Timepoint']]
duplicateIDs
```

```python
# Getting the duplicate mice by ID number that shows up for Mouse ID and Timepoint. (returning series)
duplicateIDs1 = mousedata_and_results.loc[mousedata_and_results.duplicated(subset = ['Mouse ID', 'Timepoint']),'Mouse ID'].unique()
duplicateIDs1
```

```python
# Optional: Get all the data for the duplicate mouse ID. 
duplicate_rows = mousedata_and_results.loc[mousedata_and_results['Mouse ID'] == 'g989']
print(f"Here are all the duplicate rows for mouse id: {duplicateIDs1}") 
duplicate_rows
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
mousestudy_cleaned = mousedata_and_results.loc[mousedata_and_results['Mouse ID'] != 'g989']
print(f"Here is the cleaned dataframe") 
mousestudy_cleaned
```

```python
 # Checking the number of mice in the clean DataFrame.
mousestudy_cleaned['Mouse ID'].nunique()
print(f"The number of unique mice in the cleaned dataframe is :{mousestudy_cleaned['Mouse ID'].nunique()}")
```

```python
#Summary Statistics
# Generate a summary statistics table of mean, median, variance, standard deviation, 
#and SEM of the tumor volume for each regimen
# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
# mean, median, variance, standard deviation, and SEM of the tumor volume.
mousestudy_cleaned.groupby('Drug Regimen')[["Tumor Volume (mm3)"]].mean() #--- this produces a dataframe
mousestudy_cleaned.groupby('Drug Regimen')["Tumor Volume (mm3)"].mean() #--- this produces a series. We will use series and build the DF.
```

```python
# Assemble the resulting series into a single summary dataframe.
summary1_df = pd.DataFrame({'Mean': mousestudy_cleaned.groupby('Drug Regimen')["Tumor Volume (mm3)"].mean(), 
                           'Median': mousestudy_cleaned.groupby('Drug Regimen')["Tumor Volume (mm3)"].median(), 
                           'Variance': mousestudy_cleaned.groupby('Drug Regimen')["Tumor Volume (mm3)"].var(),
                           'Std Deviation': mousestudy_cleaned.groupby('Drug Regimen')["Tumor Volume (mm3)"].std(),
                           'SEM': mousestudy_cleaned.groupby('Drug Regimen')["Tumor Volume (mm3)"].sem()
                           })
summary1_df
```

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen
# Using the aggregation method, produce the same summary statistics in a single line
summary2_df = mousestudy_cleaned.groupby('Drug Regimen').agg({"Tumor Volume (mm3)": ["mean", "median","var", "std","sem"]})
summary3_df = summary2_df.rename(columns={"mean": "Mean", "median": "Median", "var": "Variance", "std": "Std Deviation", "sem": "SEM"})
summary3_df
```

```python
#Bar and Pie Charts
# Generate a bar plot showing the total number of unique mice tested on each drug regimen using pandas.
x_values = mouse_metadata['Drug Regimen'].unique()
y_values = mouse_metadata.groupby('Drug Regimen')["Mouse ID"].nunique()
#print(f"x values are : {x_values}")
#print(f"y values are : {y_values}")

df_pandas = pd.DataFrame({'x_values':mouse_metadata['Drug Regimen'].unique(),
                   'y_values':mouse_metadata.groupby('Drug Regimen')["Mouse ID"].nunique()
                  })
ax_pandas = df_pandas.plot.bar('x_values', 'y_values', alpha=0.75, color = "green", rot=90, label = "Total number of unique mice tested on each drug regimen").legend(loc='center left',bbox_to_anchor=(1.0, 0.5))
```

```python
# Generate a bar plot showing the total number of unique mice tested on each drug regimen using pyplot.
x_axis = np.arange(len(df_pandas))
tick_locations = [value for value in x_axis]
plt.bar(df_pandas.x_values, df_pandas['y_values'], color="b",alpha=0.5, align="edge")
plt.xticks(tick_locations, df_pandas['x_values'], rotation = "vertical")
plt.title("Mice Study")
plt.xlabel("Drug Regimen")
plt.ylabel("Number of unique mice")
plt.show()
```
